```python
import matplotlib.pyplot as plt
import pandas

df = pandas.read_csv('results.csv')

collumns = df.columns.tolist()

dc1_times = df.iloc[0].values.astype(float)
dc2_times = df.iloc[1].values.astype(float)

dc2_times = [ i * 100 for i in dc2_times]

x_indexes = ['spotify_1k', 'spotify_5k', 'spotify_10k', 'spotify_20k', 'spotify_30k', 'spotify_40k']
plt.plot(x_indexes, dc1_times, marker='o', label='Better')
plt.plot(x_indexes, dc2_times, marker='o', label='Worst')
plt.title('Exemplo de Gr√°fico')
plt.xlabel('Datasets')
plt.ylabel('Time exec (s)')
plt.grid(True)
plt.legend()  # Mostra a legenda
plt.show()
```
